```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Generate random data points for two classes
np.random.seed(0)
X = np.random.randn(200, 2)
y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0)

# Train an SVM with the RBF kernel
model = SVC(kernel='rbf', C=1.0)
model.fit(X, y)

# Create a mesh to plot the decision boundary
h = .02  # step size in the mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Plot the decision boundary
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

# Plot also the training points with different markers
plt.scatter(X[y == 0, 0], X[y == 0, 1], c='blue', marker='x', label='Class 0')
plt.scatter(X[y == 1, 0], X[y == 1, 1], c='red', marker='o', label='Class 1')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title('SVM with RBF Kernel')
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the RBF kernel function
def rbf_kernel(x, y, gamma=1.0):
    return np.exp(-gamma * np.linalg.norm(x - y, axis=1)**2)

# Generate sample data points
x = np.linspace(-3, 3, 100)
data = np.array([[x_i] for x_i in x])

# Select a reference point
ref_point = np.array([[0]])

# Compute the RBF kernel values for each data point with respect to the reference point
kernel_values = rbf_kernel(data, ref_point, gamma=1.0)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, kernel_values, label="RBF Kernel")
plt.scatter(ref_point, [1], color='red', label="Reference Point")
plt.xlabel("Data Points")
plt.ylabel("Kernel Value")
plt.title("RBF Kernel Visualisation")
plt.annotate(r'$e^{-\gamma \|x - x_{ref}\|^2}$', xy=(0.5, 0.8), xycoords='axes fraction', fontsize=12)
plt.legend()
plt.show()

```
